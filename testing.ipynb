```python
df = pd.read_csv('results_1.csv')
```

```python
df["detected"] = 0
df.loc[df['prob'] >= 0.95, 'detected'] = 1
```

```python
hand = df[df['type'] == 'Non-Generated'].copy()
soft = df[df['type'] == 'Soft Watermark'].copy()
hard = df[df['type'] == 'Hard Watermark'].copy()
```

```python
# Group the dataframe by 'type' and calculate the sum of 'detected' for each group
grouped_df = df.groupby('type')['detected'].sum()

# Plot the bar graph
ax = grouped_df.plot(kind='bar')

# Add the percentages to the bars
for i, v in enumerate(grouped_df):
    ax.text(i, v + 0.5, f'({v:.1f}%)', ha='center')

# Set labels and title
plt.xlabel('Type')
plt.ylabel('Count of Detected')
plt.title('Detected Counts by Type')

# Rotate the x-axis labels
plt.xticks(rotation=45)

# Show the plot
plt.show()
```
